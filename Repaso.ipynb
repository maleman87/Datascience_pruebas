En pandas, el símbolo para la operación "y" (and) es el signo "&", mejor conocido como ampersand. Mientras que el símbolo para la operación "o" (or) es el signo "|" conocido como "pipe symbol".

```python
# uso de filtros
df[(df['diet']=='low fat')&(df['kind']=='rest')]
```

```python
# uso de filtros
df[(df.diet=='low fat')&(df['kind']=='rest')]
```

```python
# .loc es valido para filtrar también pero no es su principal función
df.loc[(df['diet']=='low fat')&(df['kind']=='rest')]
```

```python
#df.loc[condición de las filas, columna] = resultado
df.loc[(df['diet']=='low fat')&(df['kind']=='rest') , 'nueva_col'] = 1 # se cumplan las 2 condiciones
df
```

```python
df['nueva_col'].value_counts()
```

```python
df['nueva_col_2'] = 0
df.loc[(df['diet']=='low fat')|(df['kind']=='rest') , 'nueva_col_2'] = 1 # se cumpla por lo menos una condición
df['nueva_col_2'].value_counts()
```

```python
# usemos where y hagamos lo mismo que con .loc

# Crear una nueva columna 'nueva_col_3' con 0 en todas las filas
df['nueva_col_3'] = 0

# Aplicar la condición y asignar 1 a las filas que cumplen con la condición
df['nueva_col_3'] = df['nueva_col_3'].where(~((df['diet'] == 'low fat') & (df['kind'] == 'rest')), 1)
df

# where se utiliza para seleccionar elementos de un DataFrame basados en una condición booleana y reemplazar los valores que no cumplen con esa condición
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.where.html
```

```python
df['nueva_col_3'] = 0

# Aplicar la condición y asignar 1 a las filas que cumplen con la condición
df['nueva_col_3'] = df['nueva_col_3'].where(~((df['diet'] == 'low fat') & (df['kind'] == 'rest')), -999)
df
```

```python
df['nueva_col_3'].value_counts()
```

```python
print((df['nueva_col_3']== df['nueva_col']).all())
print((df['nueva_col_3']== df['nueva_col_2']).all())
```
